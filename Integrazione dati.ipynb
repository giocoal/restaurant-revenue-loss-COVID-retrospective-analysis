```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("serie-storiche-ristoranti.csv", parse_dates=["data"])
df.sort_values(by="data", inplace=True)
df
```

```python
inforistoranti = pd.read_csv("Dati/Ristoranti/RistorantiCitta.csv", sep=';')
inforistoranti
```

```python
### Unione dataset df e cittaristoranti per aggiungere colonna Location
df = df.merge(inforistoranti, left_on='id_ristorante', right_on='Codice')
df
```

```python
df.drop('Codice', axis = 1, inplace=True)
df
```

```python
### Aggiunta colonna regione e provincia in base alla Location
pavia = ['Montebello', 'Voghera', 'Stradella']

for index, row in df.iterrows():
    if row['Location'] in pavia:
        df.loc[index,'Regione'] = 'Lombardia'
        df.loc[index,'Provincia'] = 'Pavia'
    else:
        df.loc[index,'Regione'] = 'Emilia Romagna'
        df.loc[index,'Provincia'] = 'Piacenza'
        
df
```

### WEEKEND, FESTIVI, ...

```python
df['Giorno'] = df['data'].dt.day_name()
df
```

```python
### Creo variabile booleana 'Weekend'

df['Weekend'] = ((df['Giorno'] == 'Saturday') | (df['Giorno'] == 'Sunday'))
df
```

```python
### In questa lista inserisco i festivi "fissi" più Pasquetta che cambia di anno in anno. Non metto la Pasqua perchè cade
### sempre di domenica, e tutte le domeniche sono festive. Ho considerato festivo anche San Valentino, data l'importanza
### per i ristoranti
festivi = ['2018-01-01', '2018-01-06','2018-04-25', '2018-05-01', '2018-06-02', '2018-08-15', '2018-11-01', '2018-12-08',
           '2018-12-25', '2018-12-26', '2019-01-01', '2019-01-06', '2019-04-25', '2019-05-01', '2019-06-02', '2019-08-15', 
           '2019-11-01', '2019-12-08', '2019-12-25', '2019-12-26', '2020-01-01', '2020-01-06', '2020-04-25', '2020-05-01', 
           '2020-06-02', '2020-08-15', '2020-11-01', '2020-12-08', '2020-12-25', '2020-12-26', '2021-01-01', '2021-01-06', 
           '2021-04-25', '2021-05-01', '2021-06-02', '2021-08-15', '2021-11-01', '2021-12-08', '2021-12-25', '2021-12-26',
           '2022-01-01', '2022-01-06', '2022-04-25', '2018-04-02', '2019-04-22', '2020-04-13', '2021-04-05', '2022-04-18',
           '2018-02-14', '2019-02-14', '2020-02-14', '2021-02-14', '2022-02-14']

### Il patrono di Montebello si festeggia la prima domenica di ottobre (essendo domenica non specifico ulteriori date)
patrono_piacenza = ['2018-07-04', '2019-07-04', '2020-07-04', '2021-07-04']
patrono_stradella = ['2018-07-12', '2019-07-12', '2020-07-12', '2021-07-12']
patrono_voghera = ['2018-05-22', '2019-05-22', '2020-05-22', '2021-05-22']
```

```python
### Aggiungo variabile booleana 'Festivi'

df['Festivo'] = (((df['Giorno'] == 'Sunday') | (df['data'].isin(festivi))) | 
                ((df['data'].isin(patrono_piacenza)) & (df['Location'] == 'Piacenza')) |
                ((df['data'].isin(patrono_stradella)) & (df['Location'] == 'Stradella')) |
                ((df['data'].isin(patrono_voghera)) & (df['Location'] == 'Voghera')))

df
```

```python
### Aggiungo colonna "Season" usando una convenzione in base alla data

df['date_offset'] = (df.data.dt.month*100 + df.data.dt.day - 320)%1300

df['Season'] = pd.cut(df['date_offset'], [0, 300, 602, 900, 1300], 
                      labels=['Spring', 'Summer', 'Autumn', 'Winter'])

df.drop('date_offset', axis=1, inplace=True)
df
```

```python
# Per qualche motivo alle date corrispondenti al 20 marzo dei vari anni non viene correttamente fatto corrispondere il valore
# 'Spring', quindi correggo manualmente

df.loc[df['Season'].isnull(), 'Season'] = 'Spring'
```

# VACCINI

```python
vaccini = pd.read_csv("Dati/somministrazioni-vaccini.csv", parse_dates=["data"])
vaccini
```

```python
### Eliminazione colonne non necessarie (elimino anche la divisione in maschi/femmine, non credo sia rilevante)

vaccini.drop(['area', 'm', 'f', 'N1', 'N2', 'ISTAT'], axis = 1, inplace=True)
```

```python
vaccini[vaccini['reg']=='Lombardia']
```

```python
### Mantengo solo Lombardia ed Emilia Romagna

vaccini_Lom_Emil = vaccini[vaccini['reg'].isin(['Lombardia', 'Emilia-Romagna'])]
vaccini_Lom_Emil
```

```python
### Sistemazione colonne

# Dose1: numero di prime dosi effettuate
# Dose2: numero di seconde dosi effettuate
# DoseUnica: numero di somministrazioni a soggetti con infezione pregressa, che completano il ciclo vaccinale con una sola dose
# Booster1: numero di somministrazioni di dose booster
# Booster3Dosi: numero di somministrazioni di dose booster a chi ha già il ciclo vaccinale primario con 3 dosi
# Booster2: numero di somministrazioni di seconda dose booster (120 giorni dopo la prima dose booster)

vaccini_Lom_Emil.columns = ['data', 'Fornitore', 'Età', 'Dose1', 'Dose2', 'DoseUnica', 'Booster1', 'Booster3Dosi', 
                            'Booster2', 'Regione']
vaccini_Lom_Emil
```

```python
### Elimino date dopo 2022-04-30

vaccini_Lom_Emil = vaccini_Lom_Emil[vaccini_Lom_Emil['data'] < '2022-05-01']
vaccini_Lom_Emil
```

```python
### Raggruppo per data e regione (non penso che serva la divisione in fasce d'età o fornitore per la questione ristorazione)

vaccini_Lom_Emil = vaccini_Lom_Emil.groupby(['data', 'Regione'], as_index=False).sum()
vaccini_Lom_Emil
```

```python
### Separo Lombardia ed Emilia

vacciniLombardia = vaccini_Lom_Emil[vaccini_Lom_Emil['Regione']=='Lombardia']
vacciniEmilia = vaccini_Lom_Emil[vaccini_Lom_Emil['Regione']=='Emilia-Romagna']
```

```python
### Creo colonne cumulate per tutte le colonne presenti, sia per Emilia Romagna che Lombardia

colonne = ['Dose1', 'Dose2', 'DoseUnica', 'Booster1', 'Booster3Dosi', 'Booster2']

for x in colonne:
    vacciniLombardia[str(x)+'Cum'] = vacciniLombardia[x].cumsum()
    vacciniEmilia[str(x)+'Cum'] = vacciniEmilia[x].cumsum()
```

```python
vacciniEmilia.head()
```

```python
vacciniLombardia.tail()
```

```python
### Converto 'Emilia-Romagna' in 'Emilia Romagna'

vacciniEmilia.Regione = vacciniEmilia.Regione.str.replace('-', ' ')
vacciniEmilia
```

```python
vacciniLombardia.to_csv("Dati\VacciniLombardia.csv")
vacciniEmilia.to_csv("Dati\VacciniEmiliaRomagna.csv")
```

```python
vaccini_Lom_Emil = pd.concat([vacciniEmilia, vacciniLombardia])
vaccini_Lom_Emil.sort_values('data', inplace=True)
vaccini_Lom_Emil.head(20)
```

```python
vaccini_Lom_Emil.to_csv("Dati\VacciniLombardia-EmiliaRomagna.csv")
```

```python
df = pd.merge(df, vaccini_Lom_Emil,  how='left', left_on=['data','Regione'], right_on = ['data','Regione'])
df
```

# CEREALI

Tra tutti i prodotti agroalimentari essenziali per il nostro paese, il grano, e i cereali in generale, occupano un posto di grande rilievo. Buona parte della dieta italiana (e mondiale) è composta da alimenti a base di cereali quali grano, mais, olio e quant'altro. Tali prodotti sono anche estremamente importanti per la produzione di altri prodotti, basti pensare all'allevamento dove sono fondamentali come alimenti per il bestiame. L'Italia è un paese che fa grande affidamento sull'importazione di cereali dall'estero (circa il 63% del fabbisogno di grano viene importato, circa il 56% per il mais): per questo motivo risentiamo in maniera particolare delle conseguenze della guerra fra Ucraina e Russia, due dei maggiori produttori di grano a livello mondiale. Per questo motivo particolare, decidiamo di prendere in considerazione solamente i dati relativi ai prezzi dei cereali. Inoltre, pur non essendo a conoscenza della tipologia di ristoranti in questione, i cereali e i prodotti da essi derivanti sono sicuramente una parte importante del loro menù (pane, pasta, pizza, ...) mentre lo stesso non può essere affermato con altrettanta sicurezza per altri prodotti. Risultano, sul piano geopolitico, sicuramente più rilevanti i discorsi sui trasporti di tali prodotti, già toccati considerando i dati sui carburanti e sulla situazione del COVID.
