```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("serie-storiche-ristoranti.csv", parse_dates=["data"])
df.sort_values(by="data", inplace=True)
df
```

```python
df.dtypes
```

```python
rist = []
for id, df_id in df.groupby('id_ristorante', as_index=False):
    temp = df_id.reset_index()
    temp.index.name = 'rist_index'
    rist.append(temp)
    
#rist = [df_id.reset_index() for id, df_id in df.groupby('id_ristorante', as_index=False)]
```

```python
[print(df['id_ristorante'][0], df.shape) for df in rist]
```
